```python
import os
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

# Path to the directory with nii files
nii_files_path = '/content/abide_data/Outputs/cpac/nofilt_noglobal/reho'

# Output directory to save the extracted slices
output_dir = '/content/extracted_slices'

# Create output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# List all nii files
nii_files = [f for f in os.listdir(nii_files_path) if f.endswith('.nii') or f.endswith('.nii.gz')]

# Function to save slices from the 3D image
def save_slices(image_data, file_name):
    """
    Save slices from the 3D image into a folder named after the original .nii file.
    """
    # Create a folder for the current nii file
    file_output_dir = os.path.join(output_dir, file_name)
    if not os.path.exists(file_output_dir):
        os.makedirs(file_output_dir)

    # Save each slice as a PNG image
    for i in range(image_data.shape[2]):
        slice_img = image_data[:, :, i]
        output_slice_path = os.path.join(file_output_dir, f'slice_{i:03d}.png')
        plt.imsave(output_slice_path, slice_img, cmap='gray')

# Loop through each nii file
for nii_file in nii_files:
    file_path = os.path.join(nii_files_path, nii_file)

    # Load the .nii file using nibabel
    img = nib.load(file_path)

    # Get the image data as a numpy array
    image_data = img.get_fdata()

    # Remove the extension from the nii file name to use as the folder name
    file_name = os.path.splitext(nii_file)[0]

    # Save the slices to the output folder
    print(f'Extracting slices for: {nii_file}')
    save_slices(image_data, file_name)

print('Extraction and saving completed.')
```

```python
import os
import random
import shutil
from sklearn.model_selection import train_test_split

# Define paths
nii_files_path = '/content/abide_data/Outputs/cpac/nofilt_noglobal/reho'  # Path with nii files
output_dir = '/content/split_abide_data'  # Path where the split data will be stored

# Create output directories for train, validation, and test sets
train_dir = os.path.join(output_dir, 'train')
val_dir = os.path.join(output_dir, 'val')
test_dir = os.path.join(output_dir, 'test')

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Get a list of all nii files
nii_files = [f for f in os.listdir(nii_files_path) if f.endswith('.nii') or f.endswith('.nii.gz')]

# Shuffle the files
random.shuffle(nii_files)

# Split dataset: 80% train, 15% validation, 5% test
train_files, temp_files = train_test_split(nii_files, test_size=0.2, random_state=42)  # First split 80% train
val_files, test_files = train_test_split(temp_files, test_size=0.25, random_state=42)  # Then split the 20% into 15% val and 5% test

# Function to copy files to respective directories
def copy_files(file_list, destination_dir):
    for file_name in file_list:
        src_file = os.path.join(nii_files_path, file_name)
        dst_file = os.path.join(destination_dir, file_name)
        shutil.copy(src_file, dst_file)

# Copy files into their respective directories
copy_files(train_files, train_dir)
copy_files(val_files, val_dir)
copy_files(test_files, test_dir)

# Output the file counts
print(f'Total files: {len(nii_files)}')
print(f'Training set: {len(train_files)} files')
print(f'Validation set: {len(val_files)} files')
print(f'Test set: {len(test_files)} files')
```
